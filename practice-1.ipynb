````python
import re
import nltk
from nltk.corpus import stopwords
import pandas as pd
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
snowballstemmer=SnowballStemmer('english')

with open('datasets/hamlet.txt', 'r', encoding = 'utf-8') as file:
    text = file.read()
    
sentence = nltk.sent_tokenize(text)
text_df = pd.DataFrame(sentence, columns=['sentence'])
corpus = []
corpus_lemm = []

for i in range(len(sentence)):
    word = sentence[i].lower()
    word = re.sub('[^a-zA-Z]', ' ', word)
    word = nltk.word_tokenize(str(word))
    word = [snowballstemmer.stem(word) for word in word if word not in set(stopwords.words('english'))]
    word_L = [lemmatizer.lemmatize(word) for word in word if word not in set(stopwords.words('english'))]
    corpus_lemm.append(word_L)
    corpus.append(word)
````

````python
import re
import nltk
from nltk.corpus import stopwords
import pandas as pd
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
snowballstemmer=SnowballStemmer('english')

corpus = []
for i in range(len(messages)):
    word = messages["message"].iloc[i].lower()
    word = re.sub('[^a-zA-Z]', ' ', word)
    word = nltk.word_tokenize(str(word))
    word = [snowballstemmer.stem(word) for word in word if word not in set(stopwords.words('english'))]
    corpus.append(word)
````

````python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=100,binary=True)
X = cv.fit_transform(messages["message"]).toarray()
````

````python
vocab_size = len(cv.vocabulary_)
print("Vocabulary Size: ", vocab_size)
````

Why use max_features ?

- Limits dimensionality: Reduces the number of features (columns), making models faster and less memory-intensive.
- Focuses on most frequent words: Keeps only the top 100 most common words, which are likely more informative.
- Prevents overfitting: Reducing rare words/noise can help generalization.
- Useful for small datasets or quick prototyping.
However, it may exclude rare but important words. Use it when you want a compact, efficient feature set.

````python
spam_indices = [messages["label"].iloc[i]=='spam' for i in range(len(messages))] # checks each row in column label and assigns the boolean value on the basis of the given condition
spam_words = X[spam_indices].sum(axis=0) # selects the rows in the BoW matrix that are spam messages and sums each column to get total counts of each word in spam messages
spam_freq = [(word, spam_words[idx]) for word, idx in cv.vocabulary_.items()] # creates a list of tuples (word, count in spam messages)
spam_top10 = sorted(spam_freq, key=lambda x: x[1], reverse=True)[:10]
print("Top 10 words in spam messages:", [w for w, _ in spam_top10])
````

````python
ham_indices = [messages["label"].iloc[i]=='ham' for i in range(len(messages))]
ham_words = X[ham_indices].sum(axis=0)
ham_freq = [(word, ham_words[idx]) for word, idx in cv.vocabulary_.items()]
ham_top10 = sorted(ham_freq, key=lambda x: x[1], reverse=True)[:10]
print("Top 10 words in ham messages:", [w for w, _ in ham_top10])
````
